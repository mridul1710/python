```python
'''The first
function is defining the potential function whereas the second function is
objective function we will minimize using the curve_fit package'''

def lennard_jones(r, eps, sigma): #function for the lennard jones potential
    return (4 * eps) * ((sigma / r) ** 12 - (sigma / r) ** 6)
```

**I will be using the curve_fit module from the scipy package**

```python
x0 = [1, 2] #initial guesses for episolon and sigma
```

```python
#running the functions and fitting the data to find eps and sigma for He2
popt, cov = curve_fit(lennard_jones, r_val, He2_Ur, p0=x0, maxfev=10000)
eps_fit, sigma_fit = popt
print(f"Fitted parameters for He2: eps={eps_fit}, sigma={sigma_fit}")
```

```python
#running the functions and fitting the data to find eps and sigma for Ar2
params, cov = curve_fit(lennard_jones,r_val,Ar2_Ur,p0=x0, maxfev=10000)
eps_fit, sigma_fit = params
print(f"Fitted parameters for Ar2: eps={eps_fit}, sigma={sigma_fit}")
```

```python
#running the functions and fitting the data to find eps and sigma for Ne2
params, cov = curve_fit(lennard_jones,r_val,Ne2_Ur,p0=x0, maxfev=10000)
eps_fit, sigma_fit = params
print(f"Fitted parameters for Ne2: eps={eps_fit}, sigma={sigma_fit}")
```

```python
#running the functions and fitting the data to find eps and sigma for Kr2
params, cov = curve_fit(lennard_jones,r_val,Kr2_Ur,p0=x0, maxfev=10000)
eps_fit, sigma_fit = params
print(f"Fitted parameters for Kr2: eps={eps_fit}, sigma={sigma_fit}")
```

```python
'''We will pass the fitted values of eps and sigma into the original
lennard jones function so that we got values of U(r) to plot'''
He2_Ur_toplot = lennard_jones(r_val,4.754079756268886,2.4054496660441327)
Ar2_Ur_toplot = lennard_jones(r_val,13.68811715274988,3.424429990218767)
Ne2_Ur_toplot = lennard_jones(r_val,8.669434135367132,2.6845438719514507)
Kr2_Ur_toplot = lennard_jones(r_val,18.217930482718945,3.697701991563213)
```

```python
'''We will define two functions to calculate 
dU/dr and F(r) as well, F(r) = -dU/dr, as potential is a conservative force.
F(r) is 0 at minimum of U(r), this minimum indicates the point at which two molecules are
at their most stable, as learned in Assignment 1, finding minima of functions'''


def min_Ur(r,eps,sigma):
    return (24*eps*(sigma**6)*(r**-7)*((-2*(sigma**6)*r**-6)+1))  #derivative calculated analytically

'''Preparing F(r) values for all nobel gases to plot'''

F_He2 = -1*min_Ur(r_val,4.754079756268886,2.4054496660441327)
F_Ar2 = -1*min_Ur(r_val,13.68811715274988,3.424429990218767)
F_Ne2 = -1*min_Ur(r_val,8.669434135367132,2.6845438719514507)
F_Kr2 = -1*min_Ur(r_val,18.217930482718945,3.697701991563213)
         
```

```python
plt.subplots(figsize=(8, 8))
plt.plot(r_val,He2_Ur_toplot,label='Helium')
plt.plot(r_val,Ar2_Ur_toplot,label='Argon')
plt.plot(r_val,Ne2_Ur_toplot,label='Neon')
plt.plot(r_val,Kr2_Ur_toplot,label='Krypton')
plt.title('U(r) vs r for nobel gases')
plt.xlabel('r/Å')
plt.ylabel('U(r)/meV')
plt.ylim([-100,800])
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
'''Let us write code to find the minima for U(r) vs r for each nobel
gas'''

from scipy.optimize import minimize_scalar

# find the minimum of the U(r) function for all gases
He2_min_Ur = minimize_scalar(lennard_jones, args=(4.754079756268886, 2.4054496660441327), bounds=(1, 3), method='bounded')
Ar2_min_Ur = minimize_scalar(lennard_jones, args=(13.68811715274988, 3.424429990218767), bounds=(1, 3), method='bounded')
Ne2_min_Ur = minimize_scalar(lennard_jones, args=(8.669434135367132, 2.6845438719514507), bounds=(1, 3), method='bounded')
Kr2_min_Ur = minimize_scalar(lennard_jones, args=(18.217930482718945, 3.697701991563213), bounds=(1, 3), method='bounded')
```

**we will be using the minimize_scalar package from SciPy.
SciPy documentation: https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize_scalar.html**


```python
# print the minimum value and its location for all gases
print("Minimum value for U(r) for He2: ", He2_min_Ur.fun)
print("Location of minimum for He2: ", He2_min_Ur.x)

print("Minimum value for U(r) for Ar2: ", Ar2_min_Ur.fun)
print("Location of minimum for Ar2: ", Ar2_min_Ur.x)

print("Minimum value for U(r) for Ne2: ", Ne2_min_Ur.fun)
print("Location of minimum for Ne2: ", Ne2_min_Ur.x)

print("Minimum value for U(r) for Kr2: ", Kr2_min_Ur.fun)
print("Location of minimum for Kr2: ", Kr2_min_Ur.x)
```
