```python
# apply DFT, using formulas given in the assignment 4
N = len(Amp)
dft_real = np.zeros(N) #initializing arrays with the same shape as N
dft_imag = np.zeros(N)

for k in range(N):  #iterating to find the real and imaginary
    for n in range(N):
        dft_real[k] += Amp[n] * np.cos(2 * np.pi * k * n / N)
        dft_imag[k] -= Amp[n] * np.sin(2 * np.pi * k * n / N)


# plot real and imaginary parts of DFT
component_index = np.arange(N)
plt.figure()
plt.plot(component_index, dft_real, label='Real part')
plt.plot(component_index, dft_imag, label='Imaginary part')
plt.xlabel('Component index')
plt.ylabel('DFT coefficient')
plt.title('Real and Imaginary parts of DFT')
plt.legend()
plt.show()
```

**The above graph shows that the DFT has a a stronger presence of imaginary coponents**

```python
power_spectrum = (dft_real ** 2 + dft_imag ** 2) / N #calculating magnitude of the real and imaginary components


sampling_frequency = 10000 
frequency_resolution = sampling_frequency / N
peak_index = np.argmax(power_spectrum)
peak_frequency = peak_index * frequency_resolution


plt.figure()
plt.plot(component_index, power_spectrum)
plt.xlabel('Component index')
plt.xlim(0,600)
plt.ylabel('Power')
plt.title('Power spectrum')
```

**From the above figure we can observe that v1,v2,v3 lie at the component index of 100,200 and 400**

```python
'''To plot the power spectrum against frequency, you need to calculate the 
frequency values corresponding to each component index. This can be done using the following formula:
'''

def freq(comp_v):  #component index is the parameter
    return comp_v * (1000/len(Amp))
```

```python
'''Let us reconstruct the signal based on values of v1,v2,v3 
and the Fourier coefficients'''
import numpy as np

# Define the frequencies
v1 = 45.3514739229025
v2 = 90.702947845805
v3 = 181.40589569161

# Define the Fourier coefficients obtained
A = dft_real - 1j * dft_imag

# Define the length of the signal
N = len(Amp)

# Construct the frequency axis
freq_axis = np.arange(N) * sampling_frequency / N

# Find the indices corresponding to the frequencies of interest
v1_index = np.argmin(np.abs(freq_axis - v1))
v2_index = np.argmin(np.abs(freq_axis - v2))
v3_index = np.argmin(np.abs(freq_axis - v3))

# Reconstruct the signal using the frequencies and Fourier coefficients
reconstructed_signal = np.real(1/N * (Amp[0] + Amp[v1_index] * np.exp(1j * 2*np.pi * v1 * np.arange(N) / sampling_frequency) + Amp[v2_index] * np.exp(1j * 2*np.pi * v2 * np.arange(N) / sampling_frequency) + Amp[v3_index] * np.exp(1j * 2*np.pi * v3 * np.arange(N) / sampling_frequency)))

# Plot the reconstructed signal
import matplotlib.pyplot as plt
plt.figure()

plt.plot(t, reconstructed_signal, label='Reconstructed signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Reconstructed signal')

plt.show()
```
